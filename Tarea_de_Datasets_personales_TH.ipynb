```python
!pip install -q kagglehub

import kagglehub

# Descargar la versi√≥n m√°s reciente del dataset
path = kagglehub.dataset_download("wardabilal/global-music-hits-youtube-top-100-2025")

print("üìÅ Ruta de los archivos descargados:", path)

import pandas as pd
import os

# Encontrar el archivo CSV dentro del directorio descargado
for file in os.listdir(path):
    if file.endswith('.csv'):
        csv_path = os.path.join(path, file)

print("‚úÖ Archivo CSV encontrado:", csv_path)

# Cargar el dataset
df = pd.read_csv(csv_path, encoding='utf-8')
df.head()

# Vemos las primeras y √∫ltimas filas
display(df.head())
display(df.tail())

# Dimensiones (filas, columnas)
print("Shape:", df.shape)

# Informaci√≥n general del dataset
df.info()

# Estad√≠sticas descriptivas (n√∫meros y texto)
df.describe(include='all')

# Promedio de vistas
mean_views = df['view_count'].mean()
print(f"üìä Promedio de visualizaciones: {mean_views:,.0f}")

# Video con mayor y menor duraci√≥n
longest = df.loc[df['duration'].idxmax()]
shortest = df.loc[df['duration'].idxmin()]

print("\nüéµ Video m√°s largo:")
print(f"   {longest['title']} - {longest['duration']} segundos")

print("\nüéµ Video m√°s corto:")
print(f"   {shortest['title']} - {shortest['duration']} segundos")

# G√©nero m√°s frecuente
most_common_genre = df['categories'].mode()[0]
print(f"\nüé∂ Categoria m√°s popular: {most_common_genre}")

# Creamos una columna de 'Popularidad' basada en el n√∫mero de vistas
df['Popularity'] = pd.cut(
    df['view_count'],
    bins=[0, 5_000_000, 20_000_000, 100_000_000, float('inf')],
    labels=['Baja', 'Media', 'Alta', 'Muy Alta']
)

# Mostramos ejemplo
df[['title', 'view_count', 'Popularity']].head(10)
```
