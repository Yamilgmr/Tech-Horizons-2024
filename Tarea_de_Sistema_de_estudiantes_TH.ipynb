```python
estudiantes = [
    {"nombre": "Ana", "edad": 15, "promedio": 8.5, "grado": "10Â°", "materias_favoritas": ["MatemÃ¡ticas", "Arte"]},
    {"nombre": "Luis", "edad": 16, "promedio": 9.0, "grado": "11Â°", "materias_favoritas": ["Historia", "FÃ­sica"]},
    {"nombre": "MarÃ­a", "edad": 15, "promedio": 7.8, "grado": "10Â°", "materias_favoritas": ["BiologÃ­a", "QuÃ­mica"]},
    {"nombre": "Carlos", "edad": 17, "promedio": 9.2, "grado": "12Â°", "materias_favoritas": ["Deportes", "MatemÃ¡ticas"]},
    {"nombre": "SofÃ­a", "edad": 16, "promedio": 8.7, "grado": "11Â°", "materias_favoritas": ["Literatura", "InglÃ©s"]}
]

# Agregar estudiante
def agregar_estudiante(nombre, edad, promedio, grado, materias_favoritas):
    estudiante = {
        "nombre": nombre,
        "edad": edad,
        "promedio": promedio,
        "grado": grado,
        "materias_favoritas": materias_favoritas
    }
    estudiantes.append(estudiante)
    print(f"\nâœ… Estudiante {nombre} agregado correctamente.\n")

# Mostrar todos los estudiantes
def mostrar_estudiantes():
    if not estudiantes:
        print("\nâš ï¸ No hay estudiantes registrados.\n")
        return
    print("\n=== LISTA DE ESTUDIANTES ===")
    for e in estudiantes:
        print(f"ğŸ‘¤ Nombre: {e['nombre']}, Edad: {e['edad']}, Promedio: {e['promedio']}, Grado: {e['grado']}, Materias favoritas: {', '.join(e['materias_favoritas'])}")
    print()

# Buscar estudiante por nombre
def buscar_estudiante(nombre):
    for e in estudiantes:
        if e["nombre"].lower() == nombre.lower():
            print(f"\nâœ… Estudiante encontrado:")
            print(f"ğŸ‘¤ Nombre: {e['nombre']}, Edad: {e['edad']}, Promedio: {e['promedio']}, Grado: {e['grado']}, Materias favoritas: {', '.join(e['materias_favoritas'])}\n")
            return e
    print("\nâŒ Estudiante no encontrado.\n")
    return None

# Eliminar estudiante
def eliminar_estudiante(nombre):
    estudiante = buscar_estudiante(nombre)
    if estudiante:
        estudiantes.remove(estudiante)
        print(f"\nğŸ—‘ï¸ Estudiante {nombre} eliminado correctamente.\n")

# Actualizar informaciÃ³n de estudiante
def actualizar_estudiante(nombre):
    estudiante = buscar_estudiante(nombre)
    if estudiante:
        print("Deja en blanco si no deseas cambiar un valor.")
        nuevo_nombre = input("Nuevo nombre: ") or estudiante["nombre"]
        nueva_edad = input("Nueva edad: ")
        nueva_edad = int(nueva_edad) if nueva_edad else estudiante["edad"]
        nuevo_promedio = input("Nuevo promedio: ")
        nuevo_promedio = float(nuevo_promedio) if nuevo_promedio else estudiante["promedio"]
        nuevo_grado = input("Nuevo grado: ") or estudiante["grado"]
        nuevas_materias = input("Nuevas materias favoritas (separadas por coma): ")
        nuevas_materias = [m.strip() for m in nuevas_materias.split(",")] if nuevas_materias else estudiante["materias_favoritas"]

        estudiante.update({
            "nombre": nuevo_nombre,
            "edad": nueva_edad,
            "promedio": nuevo_promedio,
            "grado": nuevo_grado,
            "materias_favoritas": nuevas_materias
        })
        print(f"\nâœï¸ Estudiante {nuevo_nombre} actualizado correctamente.\n")

# Calcular estadÃ­sticas del grupo
def calcular_estadisticas():
    if not estudiantes:
        print("\nâš ï¸ No hay estudiantes para calcular estadÃ­sticas.\n")
        return
    suma = sum(e["promedio"] for e in estudiantes)
    promedio_grupo = suma / len(estudiantes)
    print(f"\nğŸ“Š Promedio general del grupo: {promedio_grupo:.2f}\n")

# MenÃº principal
def menu():
    while True:
        print("=== MENÃš PRINCIPAL ===")
        print("1ï¸âƒ£  Agregar estudiante")
        print("2ï¸âƒ£  Mostrar estudiantes")
        print("3ï¸âƒ£  Buscar estudiante")
        print("4ï¸âƒ£  Eliminar estudiante")
        print("5ï¸âƒ£  Actualizar estudiante")
        print("6ï¸âƒ£  Calcular estadÃ­sticas")
        print("7ï¸âƒ£  Salir")

        opcion = input("\nğŸ‘‰ Elige una opciÃ³n: ")

        if opcion == "1":
            nombre = input("Nombre: ")
            edad = int(input("Edad: "))
            promedio = float(input("Promedio: "))
            grado = input("Grado: ")
            materias_favoritas = input("Materias favoritas (separadas por coma): ").split(",")
            materias_favoritas = [m.strip() for m in materias_favoritas]
            agregar_estudiante(nombre, edad, promedio, grado, materias_favoritas)
        elif opcion == "2":
            mostrar_estudiantes()
        elif opcion == "3":
            nombre = input("Ingresa el nombre a buscar: ")
            buscar_estudiante(nombre)
        elif opcion == "4":
            nombre = input("Ingresa el nombre del estudiante a eliminar: ")
            eliminar_estudiante(nombre)
        elif opcion == "5":
            nombre = input("Ingresa el nombre del estudiante a actualizar: ")
            actualizar_estudiante(nombre)
        elif opcion == "6":
            calcular_estadisticas()
        elif opcion == "7":
            print("\nğŸ‘‹ Saliendo del sistema... Â¡Hasta luego!\n")
            break
        else:
            print("\nâŒ OpciÃ³n no vÃ¡lida. Intenta de nuevo.\n")

# Iniciar el programa
menu()
```
