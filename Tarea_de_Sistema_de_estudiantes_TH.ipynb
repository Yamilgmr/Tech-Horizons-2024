```python
estudiantes = [
    {"nombre": "Ana", "edad": 15, "promedio": 8.5, "grado": "10°", "materias_favoritas": ["Matemáticas", "Arte"]},
    {"nombre": "Luis", "edad": 16, "promedio": 9.0, "grado": "11°", "materias_favoritas": ["Historia", "Física"]},
    {"nombre": "María", "edad": 15, "promedio": 7.8, "grado": "10°", "materias_favoritas": ["Biología", "Química"]},
    {"nombre": "Carlos", "edad": 17, "promedio": 9.2, "grado": "12°", "materias_favoritas": ["Deportes", "Matemáticas"]},
    {"nombre": "Sofía", "edad": 16, "promedio": 8.7, "grado": "11°", "materias_favoritas": ["Literatura", "Inglés"]}
]

# Agregar estudiante
def agregar_estudiante(nombre, edad, promedio, grado, materias_favoritas):
    estudiante = {
        "nombre": nombre,
        "edad": edad,
        "promedio": promedio,
        "grado": grado,
        "materias_favoritas": materias_favoritas
    }
    estudiantes.append(estudiante)
    print(f"\n✅ Estudiante {nombre} agregado correctamente.\n")

# Mostrar todos los estudiantes
def mostrar_estudiantes():
    if not estudiantes:
        print("\n⚠️ No hay estudiantes registrados.\n")
        return
    print("\n=== LISTA DE ESTUDIANTES ===")
    for e in estudiantes:
        print(f"👤 Nombre: {e['nombre']}, Edad: {e['edad']}, Promedio: {e['promedio']}, Grado: {e['grado']}, Materias favoritas: {', '.join(e['materias_favoritas'])}")
    print()

# Buscar estudiante por nombre
def buscar_estudiante(nombre):
    for e in estudiantes:
        if e["nombre"].lower() == nombre.lower():
            print(f"\n✅ Estudiante encontrado:")
            print(f"👤 Nombre: {e['nombre']}, Edad: {e['edad']}, Promedio: {e['promedio']}, Grado: {e['grado']}, Materias favoritas: {', '.join(e['materias_favoritas'])}\n")
            return e
    print("\n❌ Estudiante no encontrado.\n")
    return None

# Eliminar estudiante
def eliminar_estudiante(nombre):
    estudiante = buscar_estudiante(nombre)
    if estudiante:
        estudiantes.remove(estudiante)
        print(f"\n🗑️ Estudiante {nombre} eliminado correctamente.\n")

# Actualizar información de estudiante
def actualizar_estudiante(nombre):
    estudiante = buscar_estudiante(nombre)
    if estudiante:
        print("Deja en blanco si no deseas cambiar un valor.")
        nuevo_nombre = input("Nuevo nombre: ") or estudiante["nombre"]
        nueva_edad = input("Nueva edad: ")
        nueva_edad = int(nueva_edad) if nueva_edad else estudiante["edad"]
        nuevo_promedio = input("Nuevo promedio: ")
        nuevo_promedio = float(nuevo_promedio) if nuevo_promedio else estudiante["promedio"]
        nuevo_grado = input("Nuevo grado: ") or estudiante["grado"]
        nuevas_materias = input("Nuevas materias favoritas (separadas por coma): ")
        nuevas_materias = [m.strip() for m in nuevas_materias.split(",")] if nuevas_materias else estudiante["materias_favoritas"]

        estudiante.update({
            "nombre": nuevo_nombre,
            "edad": nueva_edad,
            "promedio": nuevo_promedio,
            "grado": nuevo_grado,
            "materias_favoritas": nuevas_materias
        })
        print(f"\n✏️ Estudiante {nuevo_nombre} actualizado correctamente.\n")

# Calcular estadísticas del grupo
def calcular_estadisticas():
    if not estudiantes:
        print("\n⚠️ No hay estudiantes para calcular estadísticas.\n")
        return
    suma = sum(e["promedio"] for e in estudiantes)
    promedio_grupo = suma / len(estudiantes)
    print(f"\n📊 Promedio general del grupo: {promedio_grupo:.2f}\n")

# Menú principal
def menu():
    while True:
        print("=== MENÚ PRINCIPAL ===")
        print("1️⃣  Agregar estudiante")
        print("2️⃣  Mostrar estudiantes")
        print("3️⃣  Buscar estudiante")
        print("4️⃣  Eliminar estudiante")
        print("5️⃣  Actualizar estudiante")
        print("6️⃣  Calcular estadísticas")
        print("7️⃣  Salir")

        opcion = input("\n👉 Elige una opción: ")

        if opcion == "1":
            nombre = input("Nombre: ")
            edad = int(input("Edad: "))
            promedio = float(input("Promedio: "))
            grado = input("Grado: ")
            materias_favoritas = input("Materias favoritas (separadas por coma): ").split(",")
            materias_favoritas = [m.strip() for m in materias_favoritas]
            agregar_estudiante(nombre, edad, promedio, grado, materias_favoritas)
        elif opcion == "2":
            mostrar_estudiantes()
        elif opcion == "3":
            nombre = input("Ingresa el nombre a buscar: ")
            buscar_estudiante(nombre)
        elif opcion == "4":
            nombre = input("Ingresa el nombre del estudiante a eliminar: ")
            eliminar_estudiante(nombre)
        elif opcion == "5":
            nombre = input("Ingresa el nombre del estudiante a actualizar: ")
            actualizar_estudiante(nombre)
        elif opcion == "6":
            calcular_estadisticas()
        elif opcion == "7":
            print("\n👋 Saliendo del sistema... ¡Hasta luego!\n")
            break
        else:
            print("\n❌ Opción no válida. Intenta de nuevo.\n")

# Iniciar el programa
menu()
```
